````python
from dotenv import load_dotenv
import os

load_dotenv()

print("GROQ_API_KEY exists?", bool(os.getenv("GROQ_API_KEY")))
print("AGNO_API_KEY exists?", bool(os.getenv("AGNO_API_KEY")))

if not os.getenv("GROQ_API_KEY"):
    raise RuntimeError(
        "GROQ_API_KEY is missing.\n"
        "1) Copy .env.example -> .env\n"
        "2) Put your key in GROQ_API_KEY\n"
        "3) Restart the kernel and re-run"
    )
````
